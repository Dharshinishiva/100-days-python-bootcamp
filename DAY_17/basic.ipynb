```python
class user:  # thus we created a class called user
    def __init__(self, id, name):
        self.id = id
        self.name = name 
        self.followers = 0  #setting default value as 0 


user_1 = user("001", "dharshini")
print(user_1.id)
print(user_1.name)
print(user_1.followers)
```

```python
class user:  # thus we created a class called user
    def __init__(self, id, name):
        self.id = id
        self.name = name 
        self.followers = 0 
        self.following = 0 
    
    def follow(self, user):
        user.followers += 1  #user2
        self.following += 1  #user1

user_1 = user("001", "dharshini")
user_2 = user("002", "puranee")

user_1.follow(user_2) 

print(user_1.followers)
print(user_1.following) 
print(user_2.followers)
print(user_2.following) 
```
